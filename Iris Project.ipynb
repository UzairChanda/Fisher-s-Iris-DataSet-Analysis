```python

#The iris data set is a classic multivariate data set, which includes the sepal length, sepal width, petal length, 
#and petal width for hundreds of samples of three species of the iris flower.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook
```

```python
iris = pd.read_csv('Iris.csv')
iris = iris[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species']]
iris.columns = ['SepalLength','SepalWidth','PetalLength','PetalWidth','Species']
```

```python
iris.head()
```

```python
#Scatter plots shows how much one variable is affected by another or the relationship between them with the help of dots in two dimensions. 
#Scatter plots are very much like line graphs in the concept that they use horizontal and vertical axes to plot data points.
#A scatter matrix is a way of comparing each column in a DataFrame to every other column in a pairwise fashion.
#The scatter matrix creates scatter plots between the different variables and histograms along the diagonals.
#This allows us to quickly see some of the more obvious patterns in the data set.

pd.plotting.scatter_matrix(frame = iris, figsize = (7.0,7.0));
```

```python
plt.figure()
pd.plotting.parallel_coordinates(iris,'Species')
#multivariate data.pandas includes a plotting tool for creating parallel coordinates plots.
#Parallel coordinate plots are a common way of visualizing high dimensional multivariate data.
#Each variable in the data set corresponds to an equally spaced parallel vertical line. 
#The values of each variable are then connected by lines between for each individual observation.
#Coloring the lines by class, in this example the species of flower,allows the viewer to more easily see any patterns or clustering.
#looking at our iris data set, we can see that the petal length and petal width are two variables that split the different species fairly clearly. With iris virginica having the longest and widest petals. And iris setosa having the shortest and narrowest petals.
```

```python
iris.plot()
```

```python
iris.plot('SepalLength', 'PetalLength', kind = 'scatter')
plt.title('SepalLength vs PetalLength')
```

```python
iris.plot('SepalWidth', 'PetalWidth', kind = 'scatter')
plt.title('SepalWidth vs PetalWidth')
```

```python
sns.pairplot(iris , hue = 'Species' , diag_kind = 'kde');
```

```python
plt.figure(figsize = (10,8))
plt.subplot(121)
plt.title('Swarmplot vs Violinplot')
sns.swarmplot('Species', 'PetalLength', data = iris);
plt.subplot(122)
sns.violinplot('Species', 'PetalLength', data = iris);
```

```python
plt.figure(figsize = (10,8))
plt.subplot(121)
plt.title('Swarmplot vs Violinplot')
sns.swarmplot('Species', 'PetalWidth', data = iris);
plt.subplot(122)
sns.violinplot('Species', 'PetalWidth', data = iris);
```

```python
plt.figure(figsize = (10,8))
plt.subplot(121)
plt.title('Swarmplot vs Violinplot')
sns.swarmplot('Species', 'SepalWidth', data = iris);
plt.subplot(122)
sns.violinplot('Species', 'SepalWidth', data = iris);
```
